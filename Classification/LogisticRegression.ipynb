# LogisticRegression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('/content/HR_comma_sep.csv')
data
```

```python
categorical_columns=data.select_dtypes(include=['object']).columns.tolist()
categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False)
encoder
```

```python
one_hot_encoded=encoder.fit_transform(data[categorical_columns])
one_hot_df=pd.DataFrame(one_hot_encoded,columns=encoder.get_feature_names_out(categorical_columns))
one_hot_df
```

```python
df_encoded=pd.concat([data,one_hot_df],axis=1)
df_encoded
```

```python
df_encoded=df_encoded.drop(categorical_columns,axis=1)
df_encoded
```

```python
y=df_encoded['left']
x=df_encoded.drop(['left'],axis=1)
```

```python
logistic_model=LogisticRegression()
logistic_model.fit(x,y)
```

```python
x_test=np.linspace(x.min(),x.max(),100000)
y_logistic_prob=logistic_model.predict_proba(x_test)[:,1]
plt.plot(x_test,y_logistic_prob,color='blue',linewidth=2)
# y_linear_pred=linear_model.predict(x_test)
# plt.plot(x_test,y_linear_pred,color='green',linestyle='__',linewidth=2,label='Linear Regression line')
plt.xlabel('Feature')
plt.ylabel('output')
plt.title('Logistic Regression')
plt.legend()
plt.show()

```
